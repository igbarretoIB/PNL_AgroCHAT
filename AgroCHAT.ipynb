```python
class TextChunk:
  def __init__(self, start_index, end_index, text, page_number, document_name):
    self.start_index = start_index
    self.end_index = end_index
    self.text = text
    self.page_number = page_number
    self.document_name = document_name

class TextProcessor:
  def __init__(self):
        pass

  def get_pdf_text(self, pdf_docs):
    text_chunks = []

    for pdf in tqdm(pdf_docs, desc='Processando PDFs', unit='pdf'):
      pdf_reader = PdfReader(pdf)
      pdf_name = pdf
      print(pdf_name)
      for page_num, page in enumerate(pdf_reader.pages, start=1):
        page_text = page.extract_text()

        # Adicione o número da página e o nome do PDF ao texto extraído
        pdf_name = pdf
        text = f"Documento: {pdf_name}, Página: {page_num}\n{page_text}\n\n"

        # Crie um TextChunk para cada página
        chunk = TextChunk(len(text_chunks), len(text_chunks) + len(text), text, page_num, pdf_name)
        text_chunks.append(chunk)

    return text_chunks

  def get_text_chunks(self, raw_text, page_number, document_name, llm):
    chunks = []
    chunk_size =1024 # Defina o tamanho do chunk conforme necessário


    # Use o CharacterTextSplitter
    text_splitter= RecursiveCharacterTextSplitter( #create a text splitter
            separators=["\n\n", "\n", "."], #split chunks at (1) paragraph, (2) line, (3) sentence, or (4) word, in that order
            chunk_size=1024, #divide into 1000-character chunks using the separators above
            chunk_overlap=100,
            length_function= llm.get_num_tokens)

    for chunk_text in text_splitter.split_text(raw_text):
        # Adicione o número da página e o nome do documento ao final de cada chunk
        chunk_text += f"\nDocumento: {document_name}, Página: {page_number}\n"

        chunk = TextChunk(len(chunks), len(chunks) + len(chunk_text), chunk_text, page_number, document_name)

        chunks.append(chunk)

    return chunks

  def get_text_chunks_from_pdfs(self, pdf_docs, llm):
    text_chunks = []
    llm = llm
    for pdf_chunk in tqdm(self.get_pdf_text(pdf_docs), desc='Processando PDFs', unit='pdf'):
          raw_text = pdf_chunk.text
          page_number = pdf_chunk.page_number
          document_name = pdf_chunk.document_name

          chunks = self.get_text_chunks(raw_text, page_number, document_name, llm)

          for chunk in chunks:
              # Atualize o TextChunk com informações adicionais
              chunk.page_number = page_number
              chunk.document_name = document_name

          text_chunks.extend(chunks)

    return text_chunks
```

```python
def listar_arquivos_pdf(diretorio):
  arquivos_pdf = [arquivo for arquivo in os.listdir(diretorio) if arquivo.lower().endswith(".pdf")]
  return arquivos_pdf
```

```python
def get_info(model, embeddings, llm):

  # Extrair os text chunks dos documentos PDF
  diretorio_pdf = os.getcwd()
  pdf_docs = listar_arquivos_pdf(diretorio_pdf)

  text_processor = TextProcessor()
  text_chunks = text_processor.get_text_chunks_from_pdfs(pdf_docs, llm)

  text = ""
  documents = []
  # Use tqdm para criar uma barra de progresso para o loop
  for chunk in tqdm(text_chunks, desc='Processando Chunks', unit='chunk'):
    # Criar um objeto Document
    text = f'Document{chunk.document_name}, page_number:{chunk.page_number}, page_content:{chunk.text}'
    documents.append(Document(page_content=text))


  #index_from_loader = index_creator.from_documents(documents) #create an vector store index from the loaded PDF

  ffdocs = FAISS.from_documents(documents, embeddings)
  ffdocs.save_local("./faiss_index")
  print("ffdocs salvo com sucesso!")

  # Salvar informações e embeddings no ChromaDB
  vectorDB = Chroma.from_documents(documents, embedding=embeddings,  persist_directory='./vectorDB_Store')
  vectorDB.persist()
  print("vectorDB salvo com sucesso!")
  return ffdocs, vectorDB #return the index to be cached by the client app
```

```python
def generate_text(input_text, model, tokenizer, max_length=512):
    inputs = tokenizer(input_text, return_tensors="pt", max_length=max_length, truncation=True)
    outputs = model(**inputs)
    generated_text = tokenizer.decode(outputs["output_ids"][0], skip_special_tokens=True)
    return generated_text
```
